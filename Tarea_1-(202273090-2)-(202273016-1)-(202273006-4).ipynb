## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Director"

```python
%%sql
CREATE TABLE director (
    id_director INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    edad INTEGER NOT NULL,
    PRIMARY KEY (id_director)
);
```

### Tabla "Actor"

```python
%%sql
CREATE TABLE actor (
    id_actor INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    edad INTEGER NOT NULl,
    PRIMARY KEY (id_actor) 
)
```

### Tabla "Pelicula"

```python
%%sql
CREATE TABLE pelicula (
    id_pelicula INTEGER GENERATED ALWAYS AS IDENTITY,
    id_director INTEGER NOT NULL,
    nombre VARCHAR(45) NOT NULL,
    genero VARCHAR(45) NOT NULL,
    duracion INTEGER NOT NULL,
    fecha_publicacion DATE NOT NULL,
    PRIMARY KEY (id_pelicula),
    FOREIGN KEY(id_director) REFERENCES director(id_director)
);
```

### Tabla "actor_participa_en_pelicula"

```python
%%sql
CREATE TABLE actor_participa_en_pelicula (
    id_actor INTEGER,
    id_pelicula INTEGER,
    FOREIGN KEY (id_pelicula) REFERENCES Pelicula (id_pelicula),
    FOREIGN KEY (id_actor) REFERENCES actor (id_actor)
);
```

### Tabla  "Premio_mejor_director"

```python
%%sql
CREATE TABLE premio_mejor_director (
    id_premio INTEGER GENERATED ALWAYS AS IDENTITY,
    id_director INTEGER,
    anio INTEGER NOT NULL,
    PRIMARY KEY (id_premio),
    FOREIGN KEY (id_director) REFERENCES director (id_director)
);
```

### Tabla  "Premio_mejor_actor"

```python
%%sql
CREATE TABLE premio_mejor_actor (
    id_premio INTEGER GENERATED ALWAYS AS IDENTITY,
    id_actor INTEGER,
    anio INTEGER NOT NULL,
    PRIMARY KEY (id_premio),
    FOREIGN KEY (id_actor) REFERENCES actor (id_actor)
);
```

### Tabla  "Premio_mejor_pelicula"

```python
%%sql
CREATE TABLE premio_mejor_pelicula (
    id_premio INTEGER GENERATED ALWAYS AS IDENTITY,
    id_pelicula INTEGER,
    anio INTEGER NOT NULL,
    PRIMARY KEY (id_premio),
    FOREIGN KEY(id_pelicula) REFERENCES pelicula(id_pelicula)
);
```

**Solución:** --se crea una tabla que seleccionara los nombres de los actores de la tabla actor, siempre  cuando el actor haya participado en una pelicula de el genero Comedia. Tambien se creara una relacion entre pelicula, actor_participa_en_pelicula y actor.

### Consulta 4
**Nombre y cantidad de premios ganados por cada género, ordenados de mayor a menor según la cantidad de premios.**


**Ejemplo:**
|Género  |Cantidad de premios|
|--------|-------------------|

**Solución:** se creara una tabla que selecciona el genero y la cantidad de premios que gano cada uno de la tabla pelicula. Se creara una conexion entre la tabla premio_mejor_pelicula con pelicula. Luego agrupamos por el genero y ordenamos de mayor a menor segun su cantidad de premios.

```python
%%sql
SELECT Pelicula.genero AS "Género", COUNT(Premio_mejor_pelicula.id_premio) AS "Cantidad de premios"
FROM Pelicula
JOIN Premio_mejor_pelicula ON Pelicula.id_pelicula = Premio_mejor_pelicula.id_pelicula
GROUP BY "Género"
ORDER BY "Cantidad de premios" DESC;
```

### Consulta 5
**Mostrar peliculas que tengan 2 o mas actores/actrices ordenadas de mayor a menor segun la cantidad de actores/actrices.**


**Ejemplo:**
|Nombre     |Cantidad de actores|
|-----------|-------------------|
|Oppenheimer|7                  |

**Solución:**se creara una tabla que selecciona el nombre de la pelicula y la cantidad de actores de la tabla pelicula siempre que tenga 2 o mas de ellos. Luego creamos un join que relaciona actor_participa_en_pelicula con pelicula y luego se agruparan y ordenaran de mayor a menor.

```python
%%sql
SELECT Pelicula.nombre AS "Nombre", COUNT(Actor_participa_en_pelicula.id_actor) AS "Cantidad de actores"
FROM Pelicula
JOIN Actor_participa_en_pelicula ON Pelicula.id_pelicula = Actor_participa_en_pelicula.id_pelicula
GROUP BY Pelicula.nombre
HAVING COUNT(Actor_participa_en_pelicula.id_actor) >= 2
ORDER BY "Cantidad de actores" DESC;
```

### Consulta 6
**Nombre de la película con la mayor brecha de edad entre el actor más joven y el actor más viejo que participen en ella, y las edades de dichos actores.** (Si hay empate, mostrar solamente uno).


**Ejemplo:**
|Nombre|Edad del actor más joven|Edad del actor más viejo|
|------|------------------------|------------------------|
|Barbie|16                      |92                      |


**Solución:** se creara una tabla que seleccionara el nombre de las peliculas y buscara las edades de sus actores de la tabla pelicula. Tambien creara join´s entre actor_participa_en_pelicula, pelicula y actor, y agrupara las peliculas y las ordenara segun la mayor brecha de edad.

```python
%%sql
SELECT Pelicula.nombre AS "Nombre", 
    MIN(Actor.edad) AS "Edad del actor más joven",
    MAX(Actor.edad) AS "Edad del actor más viejo"
FROM Pelicula
JOIN Actor_participa_en_pelicula ON Pelicula.id_pelicula = Actor_participa_en_pelicula.id_pelicula
JOIN Actor ON Actor_participa_en_pelicula.id_actor = Actor.id_actor
GROUP BY Pelicula.nombre
ORDER BY MAX(Actor.edad) - MIN(Actor.edad) DESC
LIMIT 1;
```

### Consulta 7
**Año de cada premio ganado por la película de mayor duración.** (Si hay empate, elegir solamente uno).


**Ejemplo:**
|Año |
|----|
|2024|
|... |


**Solución:** se crea una tabla con el año de cada premio ganado por la pelicula de mayor duracion. Se crea un join entre las tablas pelicula y premio_mejor_pelicula. Se reconocen las peliculas de mayor duracion con SELECT(MAX()) de la tabla pelicula.

```python
%%sql
SELECT premio_mejor_pelicula.anio AS "Año"
FROM pelicula
JOIN premio_mejor_pelicula ON pelicula.id_pelicula = premio_mejor_pelicula.id_pelicula
WHERE duracion = (
    SELECT MAX(duracion)
    FROM pelicula
)
LIMIT 1;
```

### Consulta 8
**Nombre de todos los actores que han trabajado con más de un director, la cantidad de directores con las que han trabajado y el nombre de una de las películas.**


**Ejemplo:**
|Nombre del actor|Cantidad de directores|Nombre de una película|
|----------------|----------------------|----------------------|
|Cillian Murphy  |29                    |Oppenheimer           |
|...             |...                   |...                   |


**Solución:** se creara una tabla con los nombres de los actores, la cantidad de directores con los que ha trabajado y el nombre de una de sus peliculas de la tabla Actor. Luego hacemos join´s entre actor, actor_participa_en_pelicula y pelicula mediante sus id´s. Tambien agrupamos por el nombre del actor, utilizamos DISTINCT para que no se repitan los elementos y COUNT para contar los directores.

```python
%%sql
SELECT 
    Actor.nombre AS "Nombre del actor",
    COUNT(DISTINCT Pelicula.id_director) AS "Cantidad de directores",
    MIN(Pelicula.nombre) AS "Nombre de una pelicula"
FROM Actor
JOIN 
    Actor_participa_en_pelicula ON Actor.id_actor = Actor_participa_en_pelicula.id_actor
JOIN 
    Pelicula ON Actor_participa_en_pelicula.id_pelicula = Pelicula.id_pelicula
GROUP BY 
    Actor.nombre
HAVING 
    COUNT(DISTINCT Pelicula.id_director) > 1;
```

**Solución:** se creara una tabla con el nombre del actor, del director, el nombre y numero de las peliculas, siempre y cuando el actor no este en la tabla premio_mejor_actor. Haremos join´s entre actor_participa_en_pelicula, actor, pelicula y director.

```python
%%sql
SELECT 
    Actor.nombre AS "Nombre del actor",
    Pelicula.nombre AS "Nombre de la pelicula",
    Director.nombre AS "Nombre del director",
    COUNT(DISTINCT Pelicula.id_pelicula) AS "Número de películas del director",
    COUNT(DISTINCT PremioDirector.id_premio) AS "Número de premios del director"
FROM 
    Actor
JOIN 
    Actor_participa_en_pelicula ON Actor.id_actor = Actor_participa_en_pelicula.id_actor
JOIN 
    Pelicula ON Actor_participa_en_pelicula.id_pelicula = Pelicula.id_pelicula
JOIN 
    Director ON Pelicula.id_director = Director.id_director
LEFT JOIN 
    Premio_mejor_director AS PremioDirector ON Director.id_director = PremioDirector.id_director
WHERE 
    Actor.id_actor NOT IN (SELECT id_actor FROM Premio_mejor_actor)
GROUP BY 
    Actor.nombre, Pelicula.nombre, Director.nombre;
```
